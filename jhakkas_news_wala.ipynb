### **User Data**

```python
from faker import Faker
import random
import json

# Create a Faker instance with English locale
fake = Faker('en_US')


# Generate a user profile
def generate_user_profile(num_profiles):
  profiles=[]
  for _ in range(num_profiles):
      user_profile = {
          'user_id': _,
          'name': fake.name(),
          'email': fake.email(),
          'interested_categories': random.sample(['Technology', 'Health', 'Science', 'Business', 'Entertainment', 'Sports'], k=3)
      }
      profiles.append(user_profile)
  return profiles

# Generate fake data

user_profile = generate_user_profile(30)

```

```python
user_df=pd.DataFrame(user_profile)
user_df.shape
```

### **Content Based Recommendation**

```python
!pip install bertopic
!pip install sentence-transformers
!pip install scikit-learn
```

```python
from sentence_transformers import SentenceTransformer
import torch
import torch.nn.functional as F

# Load the pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Function to get sentence embeddings
def get_sentence_embedding(sentence):
    return model.encode(sentence, convert_to_tensor=True)
```

```python
news_emb=[]
for i in news_df['text']:
  news_emb.append( get_sentence_embedding(i))
```

```python
user_emb=[]
for i in user_df['interested_categories']:
  user_emb.append( get_sentence_embedding(i))
```

```python
import pandas as pd
import torch
import torch.nn.functional as F
```

```python
result={}
for i in range(0,len(user_emb)):
  scores_lst=[]
  for j in range(0,len(news_emb)):
    cosine_sim = F.cosine_similarity(user_emb[i], news_emb[j], dim=0)
    score = cosine_sim.mean().item()
    scores_lst.append([score,j])

  scores_lst=sorted(scores_lst, key=lambda x: x[0] , reverse=True)
  result[i]= scores_lst[:10]



```

```python
user_df.head()
```

```python
rec=[]
for i in result:
  ans=[]
  for j in result[i]:
    ans.append(news_df.iloc[j[1]]['text'])
    ans.append(URL_DF.iloc[j[1]]['url'])
  rec.append(ans)
rec_df=pd.DataFrame(rec,columns=[
    'Article 1', 'URL Article 1',
    'Article 2', 'URL Article 2',
    'Article 3', 'URL Article 3',
    'Article 4', 'URL Article 4',
    'Article 5', 'URL Article 5',
    'Article 6', 'URL Article 6',
    'Article 7', 'URL Article 7',
    'Article 8', 'URL Article 8',
    'Article 9', 'URL Article 9',
    'Article 10', 'URL Article 10'
])
rec_df.head()
```

```python
recphase1_df=pd.concat([user_df,rec_df],axis=1)
recphase1_df.head()
```

### **Collaborative Recommendation**

```python
from faker import Faker
import random

# Initialize the Faker instance
fake = Faker()

# Define the number of records you want to generate
num_records = 10

# Define a function to generate the dataset
def generate_dataset(num_records):
    dataset = []
    for _ in range(num_records):
        user_id = f'U{_}'
        article_id = f'A{_}'
        article_rank = random.randint(1, 581)
        time_spent = random.randint(10, 600)  # Time spent in seconds, from 1 second to 10 minutes

        record = {
            'User ID': user_id,
            'Article ID': article_id,
            'Article Rank': article_rank,
            'Time Spent': time_spent
        }

        dataset.append(record)

    return dataset

# Generate the dataset
dataset = generate_dataset(num_records)

# Print the generated dataset
for record in dataset:
    print(record)
```
